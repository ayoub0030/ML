```python
# ðŸ“¦ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ§® Settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)



# ðŸ–¼ï¸ Step 3: Visualizations

## 1ï¸âƒ£ Histograms for all numeric columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_cols].hist(bins=20, figsize=(12, 8), edgecolor='black')
plt.suptitle("ðŸ“ˆ Distribution of Numeric Features")
plt.show()

## 2ï¸âƒ£ Correlation Heatmap
if len(numeric_cols) > 1:
    plt.figure(figsize=(10, 6))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("ðŸ”¥ Correlation Heatmap")
    plt.show()

## 3ï¸âƒ£ Boxplots (for outlier detection)
for col in numeric_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"ðŸ“¦ Boxplot of {col}")
    plt.show()

## 4ï¸âƒ£ Pairplot (only if not too many columns)
if len(numeric_cols) <= 6:
    sns.pairplot(df[numeric_cols])
    plt.suptitle("ðŸ”— Pairwise Relationships")
    plt.show()

# 5ï¸âƒ£ Categorical vs Numeric (optional)
categorical_cols = df.select_dtypes(include=["object", "category"]).columns
if len(categorical_cols) > 0:
    for cat in categorical_cols:
        for num in numeric_cols:
            plt.figure()
            sns.boxplot(x=cat, y=num, data=df)
            plt.title(f"ðŸ“Š {num} by {cat}")
            plt.xticks(rotation=45)
            plt.show()


# Compute correlation matrix
correlation_matrix = df.corr()

# Visualize using a heatmap
plt.figure(figsize=(6,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()


print("\nâœ… Visualization complete!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# Load the dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target  # 0 = malignant, 1 = benign

# Split the data based on target
df_0 = df[df['target'] == 0]  # malignant
df_1 = df[df['target'] == 1]  # benign

# Display some info
print("âœ… Dataset loaded:", df.shape)
print("Target labels:", dict(enumerate(data.target_names)))

# Choose all feature columns
features = df.columns[:-1]

# Define grid for histograms
num_cols = 4
num_rows = int(np.ceil(len(features) / num_cols))
plt.figure(figsize=(20, num_rows * 3))

# Plot each feature
for i, feature in enumerate(features, 1):
    plt.subplot(num_rows, num_cols, i)
    
    plt.hist(df_0[feature], bins=25, alpha=0.6, color='red', label='Malignant (0)')
    plt.hist(df_1[feature], bins=25, alpha=0.6, color='green', label='Benign (1)')
    
    plt.title(feature)
    plt.xlabel('')
    plt.ylabel('')
    plt.legend(fontsize=8)

plt.tight_layout()
plt.show()
```
